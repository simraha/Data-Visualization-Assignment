<p>Electricity generation and electricity capacity are two different but related concepts in the energy sector:
    <ul>
    <li>Electricity generation refers to the actual production of electricity by power plants or other facilities. This is sometimes measured in units like gigawatt-hours (GW.h), which represent the amount of electricity produced over a certain period of time</li>
    <li>Electricity capacity, on the other hand, refers to the maximum amount of electricity that a power plant or facility is capable of producing under ideal conditions. This is usually measured in units like megawatts (MW), which represent the power output of the facility at a given moment</li>
</ul></p>
<p>An electricity generation and capacity dataset provides valuable insights into the functioning of the power system and can help inform policy and investment decisions related to energy infrastructure and sustainability as well as energy planning.</p>

<h2>Visualization</h2>
<p>In the graphs below I will be analyzing the data from the datasets</p>

```python
import pandas as pd
import matplotlib.pyplot as plt

#load the data from CSV file
egd = pd.read_csv('electricity-generation-dataset.csv')

#pivot the data to create a DataFrame with columns for each source
pivot_ecd = egd.pivot_table(index='Year', columns='Source', values='Data')

#plot the data as a line chart with a legend for each source
plot = pivot_ecd.plot(kind='line', figsize=(15, 8))

#set the axis labels and title
plot.set_xlabel('Year')
plot.set_ylabel('Electricity Generation (GW.h)')
plot.set_title('Electricity Generation by Source over Time')
plot.grid()

#display the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#read in the data
egd = pd.read_csv('electricity-generation-dataset.csv')

#calculate the total electricity generation by region
total_by_region = egd.groupby('Region')['Data'].sum()

#remove following from the total_by_region DataFrame
total_by_region = total_by_region.drop('Canada')
total_by_region = total_by_region.drop('YT')
total_by_region = total_by_region.drop('NT')
total_by_region = total_by_region.drop('NU')

#create the bar chart
total_by_region.plot(kind='bar', figsize=(8,6))

#set the axis labels and title
plt.xlabel('Region (province)')
plt.ylabel('Total Electricity Generation (GW.h)')
plt.title('Total Electricity Generation by Region')
plt.grid()

#display the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#load the data from CSV file
ecd = pd.read_csv('electricity-capacity-dataset.csv')

#pivot the data to create a DataFrame with columns for each source
pivot_ecd = ecd.pivot_table(index='Year', columns='Source', values='Data')

#plot the data as a line chart with a legend for each source
plot = pivot_ecd.plot(kind='line', figsize=(15, 8))

#set the axis labels and title
plot.set_xlabel('Year')
plot.set_ylabel('Electricity Capacity (MW)')
plot.set_title('Electricity Capacity by Source over Time')
plot.grid()

#display the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#read in the data
ecd = pd.read_csv('electricity-capacity-dataset.csv')

#calculate the total electricity generation by region
total_by_region = ecd.groupby('Region')['Data'].sum()

#remove Canada from the total_by_region DataFrame
total_by_region = total_by_region.drop('Canada')
total_by_region = total_by_region.drop('YT')
total_by_region = total_by_region.drop('NT')
total_by_region = total_by_region.drop('NU')

#create the bar chart
total_by_region.plot(kind='bar', figsize=(8,6))

#set the axis labels and title
plt.xlabel('Region (province)')
plt.ylabel('Total Electricity Capacity (MW)')
plt.title('Total Electricity Capacity by Region')
plt.grid()

#display the chart
plt.show()
```
